```python
sampe_data.drop(labels=["Country","Postal Code"],axis=1,inplace=True)
```

### Data Visualization

```python
sampe_data.groupby('Ship Mode')['Profit','Sales'].sum().plot.bar()
plt.title('Total Profit and Sales per Ship Mode')
plt.show()
```

##### Most of Sales sent by Standard Class Shipment

```python
sampe_data.groupby('Segment')['Profit','Sales'].sum().plot.bar()
plt.title('Total Profit and Sales per Segment')
plt.show()
```

###### Among Three Segmanent, Consumer sales are higher

```python
sampe_data.groupby('State')['Profit','Sales'].sum().plot.bar(figsize=(20,5))
plt.title('Total Profit and Sales per State')
plt.show()
```

###### 'California' and 'New York' state has high sales and profit
###### ''Colorado', 'Florida', 'illinois', 'North Carolina, 'Chio' , 'Pennslvania', 'Tennessee' and 'Texas' state have negative profit

```python
sampe_data.groupby(['Region'])['Profit','Sales'].sum().plot.bar(figsize=(10,5))
plt.title('Total Profit and Sales per Region')
plt.show()
```

```python
sampe_data.groupby(['Region','Sub-Category'])['Profit','Sales'].sum().plot.bar(figsize=(20,20))
plt.title('Total Profit and Sales per Region')
plt.show()
```

###### 'East' and 'West' Regions have more sales
###### in Central region sales, Appliances,Binder,Chairs, Furnishings,Machines,supplies and Tables have negative profit
###### in East region sales, Bookcase,Supplies and Tables have negative profit
###### in South region sales, Machines and Tables sales have negative profit
###### in West region sales, Bookcase and Machines have negative profit

```python
sampe_data.groupby('Category')['Profit','Sales'].sum().plot.bar()
plt.title('Total Profit and Sales per Category')
plt.show()
```

```python
plt.figure(figsize=(20,8))
sns.countplot(x="Category",hue="Region",data=sampe_data)
plt.show()
```

###### Sales of Office Supplies are more in Western, Central and Eastern Region

```python
sampe_data.groupby(['Category','Sub-Category'])['Profit','Sales'].sum().plot.bar(figsize=(20,10))
plt.title('Total Profit and Sales per Region')
plt.show()
```

###### Most of Profit comes from 'Office Supplies' and 'Technology'  catogary. wheater furniture has good sales but less profit
###### In Furniture category, Tables have negative profit and Chairs have highest profit
###### In Office category, Papers have highest profit
###### In Technology category, Copiers have highest profit

```python
plt.figure(figsize=(12,10))
sampe_data['Sub-Category'].value_counts().plot.pie(autopct="%1.1f%%")
plt.show()
```

###### The store has wide variety of Binders and Paper department.

```python
sampe_data.groupby('Sub-Category')['Profit','Sales'].sum().plot.bar(figsize=(15,5))
plt.title('Total Profit and Sales per Sub-Category')
plt.show()
```

###### Most of sales of 'Chair' and ''Phones' 
###### Profit of 'copies','Phones','Paper' and 'Accessories' are high wheater profit on 'Tables' are in negative

```python
plt.figure(figsize=(20,8))
sns.countplot(x="Sub-Category",hue="Region",data=sampe_data)
plt.show()
```

###### People of Western part tend to order more from superstore.
